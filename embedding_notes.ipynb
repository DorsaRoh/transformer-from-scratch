# Embeddings

Embeddings are representations of values or objects like text, images, and audio that are designed to be consumed by machine learning models and semantic search algorithms.

They represent words, phrases, or even entire sentences as dense vectors of numbers in a high-dimensional space. These numbers represent the papers in n directions (for how many numbers are in the vector for the specific embedding). They are coordinates in an n dimensional plane, where embeddings with similar significance/meaning are closer to each other in coordinates and hold closer numbers in the vector

### Dot Product

Measures how aligned two vectors/embeddings are
- 0, if perpendicular
- negative if point in opposite directions

### Normalization

Normalization usually means adjusting values to a common scale. In the context of embeddings, it often means making each vector have a length (magnitude) of 1 while preserving the direction
<br>

Why normalize?
- It makes comparing embeddings easier, as they're all on the same scale
- It focuses on the direction of the vector rather than its magnitude, which often better captures the meaning

### Cosine Similarity

`cosine_similarity` is a measure of similarity between two vectors of an inner product space (the dot product / product of lengths)

- measures the cosine of the angle between the two vectors
- for embeddings:
    - if two embeddings are similar, the cosine similarity is close to 1 (0 degrees. parallel vectors)
    - if they are very different, it will be close to 0 (90 degrees. perpendicular vectors)
    - if they are opposite, it will be close to -1 (rare with embeddings) (180 degrees. anti-parallel: parallel in opposite directions)

## Embeddings Beyond Words

For transformers, vectors in embedding spaces don't merely represent words. They represent **position** of words, and have the capcity to soak contexts. 

Ex. A vector for *"king"* can be tugged and pulled by various blocks in the network, so it points in a more specific and nuanced position, so its ex. a king that lived in Scotland, who achieved his post by murdering the previous king, who spoke in Shakespeare.
